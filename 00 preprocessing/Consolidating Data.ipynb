```python
import pandas as pd
from pathlib import Path
```

```python
RAW_DATA_DIR = Path('../raw_data/')
DATA_DIR = Path('../data/')
```

```python
# list here only the datasets we want to use in the modelling part
# make sure they have a 'ZIP Code' column

datasets = [
    'ca_filtered_2015_2025.csv',
    'cleaned_climate_disasters.csv',
    'fair_plan.csv',
    'fair_plan2022.csv',
    'percent_white_population_yearly.csv',
    'premium_change2021.csv',
    'premiums_totals.csv',
    'renewals_totals.csv',
    'yearly_lowest_income_by_zip.csv',
    'yearly_median_income_by_zip.csv',
    'zillow_CA_df_filtered_agg.csv',
]
```

```python
# loading and joining all datasets
df = pd.read_csv(DATA_DIR / datasets[0], index_col=0)

for dataset in datasets[0:]:
    print(f'Loading and merging: {dataset}')
    df = df.merge(pd.read_csv(DATA_DIR / dataset, index_col=0), on="ZIP Code")
```
