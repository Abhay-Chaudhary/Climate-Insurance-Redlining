```python
import pandas as pd
from pathlib import Path
```

```python
RAW_DATA_DIR = Path('../raw_data/')
DATA_DIR = Path('../data/')
```

```python
# list here only the datasets we want to use in the modelling part
# use tuple with (Path, name of zipcode column in dataset)

datasets = [
    (DATA_DIR / '2018_2021' / 'disaster_trends_by_zip_2022.csv', 'ZIP'),
    (DATA_DIR / '2018_2021' / 'fair_plan2022.csv', 'ZIP Code'),
    (DATA_DIR / '2018_2021' / 'median_incomes_flat.csv', 'ZIP'),
    (DATA_DIR / '2018_2021' / 'percent_white_population_flat.csv', 'ZIP'),
    (DATA_DIR / '2018_2021' / 'policy_renewals.csv', 'ZIP Code'),
    (DATA_DIR / '2018_2021' / 'premiums_claims.csv', 'ZIP Code'),
    (DATA_DIR / '2018_2021' / 'Zillow_df_filtered.csv', 'zipcode')
    (DATA_DIR / '2018_2021' / 'housing21.csv', 'zipcode')
]

datasets
```

```python
# loading all datasets

renewals = pd.read_csv(DATA_DIR / '2018_2021' / 'policy_renewals.csv', index_col=0)
premiums = pd.read_csv(DATA_DIR / '2018_2021' / 'premiums_claims.csv', index_col=0)
fair = pd.read_csv(DATA_DIR / '2018_2021' / 'fair_plan2022.csv', index_col=0)
income = pd.read_csv(DATA_DIR / '2018_2021' / 'median_incomes_flat.csv', index_col=0).rename(columns={'ZIP': 'ZIP Code'})
race = pd.read_csv(DATA_DIR / '2018_2021' / 'percent_white_population_flat.csv', index_col=0).rename(columns={'ZIP': 'ZIP Code'})
housing = pd.read_csv(DATA_DIR / '2018_2021' / 'Zillow_df_filtered.csv', index_col=0).reset_index().rename(columns={'zipcode': 'ZIP Code'})
disasters = pd.read_csv(DATA_DIR / '2018_2021' / 'disaster_summary_by_zip.csv').rename(columns={'ZIP': 'ZIP Code'})
housing2 = pd.read_csv(DATA_DIR / '2018_2021' / 'housing21.csv')
```

```python
df = pd.merge(premiums, renewals, on='ZIP Code')
df = pd.merge(df, fair, on='ZIP Code')
df = pd.merge(df, income, on='ZIP Code')
df = pd.merge(df, race, on='ZIP Code')
df = pd.merge(df, housing, on='ZIP Code')
df = pd.merge(df, housing2, on='ZIP Code')
df = pd.merge(df, disasters, on='ZIP Code', how='left')

df
```

```python
# filling data related to disaster to 0
fields = ['Count_2022', 'Sum_1yr', 'Avg_3yr', 'Sum_3yr', 'Avg_5yr', 'Sum_5yr', 'Avg_10yr', 'Sum_10yr']
df[fields] = df[fields].fillna('0')
```
