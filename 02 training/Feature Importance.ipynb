```python
# Correlations
```

```python
from scipy.stats import pearsonr

# correlations with pct_fair_plan

corrs = []

for col  in df.columns[6:]:
    corr_coeff, p_value = pearsonr(df['pct_fair_plan'], df[col])    
    corr_coeff2, p_value2 = pearsonr(df['exposure_22'], df[col])    
    corr_coeff3, p_value3 = pearsonr(np.log(df['exposure_22']), df[col])
    corrs.append([col, corr_coeff, p_value, corr_coeff2, p_value2, corr_coeff3, p_value3])

```

```python
results = pd.DataFrame(corrs, columns=['Feature', 'Corr FAIR Plan', 'p-value FAIR Plan',
                            'Corr Exposure', 'p-value Exposure',
                            'Corr Log-Exposure', 'p-value Log-Exposure']).sort_values(by='Corr FAIR Plan')
```

## Feature Importance

```python
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(
    df,
    test_size=.2,
    random_state=13348,
    shuffle=True
)

features = df.columns[6:]
```

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor(n_estimators=500, max_depth=5)

forest.fit(df_train[features], np.log(df_train['exposure_22']))
```

```python
importance = pd.DataFrame({
    'feature': features,
    'importance': forest.feature_importances_
})#.sort_values(by='importance')

importance
```
